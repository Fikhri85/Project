```python
# To import and install necessary library
!pip install yfinance==0.1.67
!mamba install bs4==4.10.0 -y
!pip install nbformat==4.2.0
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# to get share data from yfinance
man_utd=yf.Ticker('MANU')
```

The dividends of the Manchester United share can be seen as below:

```python
man_utd.dividends
```

The dividends data can be plot into graph:

```python
man_utd.dividends.plot()
```

This graph shows the increase in dividends after 2016

We can see the price of the open shares during the years:

```python
man_utd_share_data=man_utd.history(period='max')
man_utd_share_data
```

We can see the market trend using the plot function:

```python
man_utd_share_data.reset_index(inplace=True)
man_utd_share_data.plot(x='Date',y='Open')
```

We can see the rise of the share price during 2023 due the rumors of Sheikh Jassim Bin Hamad Al Thani takeover of Manchester United FC.
But the trends slowly going down after the dismissal of the rumor and the takeover by Sir Jim Ratcliffe.


Let us compare the data with its rival club; Manchester City FC, using Man City Fan Token value.

```python
man_city=yf.Ticker('CITY10049-USD')
man_city_data=man_city.history(period='max')
man_city_data
```

Let us see the trend using the plot function:

```python
man_city_data.reset_index(inplace=True)
man_city_data.plot(x='Date',y='Open')
```

We can see the downward trends of the open price which can be due to the increments of the fanbase and the status of the clubs after their recent achievement in winning competitions.

We can do comparison of the open share price between these two football clubs using the same graph. This data might have discrepencies since Man City price are referred to the Fan Token price. 

```python
import matplotlib.pyplot as plt

# Assuming man_utd_share_data and man_city_data have 'Date' and 'Open' columns
plt.figure(figsize=(10, 6))

# Plot Manchester United data
plt.plot(man_utd_share_data['Date'], man_utd_share_data['Open'], label='Manchester United', marker='o')

# Plot Manchester City data
plt.plot(man_city_data['Date'], man_city_data['Open'], label='Manchester City', marker='o')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.title('Comparison of Open Prices: Manchester United vs. Manchester City')

# Show legend
plt.legend()

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
